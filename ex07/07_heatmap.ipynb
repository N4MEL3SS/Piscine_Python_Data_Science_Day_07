## Import

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
db_connect = sqlite3.connect('../data/checking-logs.sqlite.sqlite')
```

## Analyze only the users and not the admins

```python
sql_query = """
SELECT uid, timestamp
FROM checker
WHERE uid LIKE 'user_%'
"""

commits = pd.read_sql(sql=sql_query, con=db_connect, parse_dates=['timestamp'])
commits.head()
```

```python
commits.info()
```

```python
days = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}

commits['weekday'] = commits['timestamp'].apply(lambda x: days[x.dayofweek])
commits['hour'] = commits['timestamp'].dt.hour
commits = commits.drop('timestamp', axis=1)
commits.head()
```

```python
commits.info()
```

```python
columns = commits.groupby('uid').count().sort_values('hour', ascending=False).index
commits_day = commits.groupby('uid').weekday.value_counts().unstack(0)
commits_day = commits_day.reindex(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
commits_day = commits_day.reindex(columns=columns)
commits_day = commits_day.fillna(0)
commits_day
```

```python
index = pd.RangeIndex(stop=24)
commit_hour = commits.groupby('uid').hour.value_counts().unstack(0)
commit_hour = commit_hour.reindex(index=index, fill_value=0)
commit_hour = commit_hour.reindex(columns=columns)
commit_hour = commit_hour.rename_axis('hour')
commit_hour = commit_hour.fillna(0)
commit_hour
```

```python
fig_1, ax = plt.subplots(figsize=(15, 8))
im = ax.imshow(commits_day, cmap='RdBu')
divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='5%', pad=0.05)
plt.colorbar(im, cax=cax)

ax.set_xticks(range(commits_day.shape[1]))
ax.set_yticks(range(commits_day.shape[0]))
ax.set_xticklabels(commits_day.columns)
ax.set_yticklabels(commits_day.index)

plt.setp(ax.get_xticklabels(), rotation=90)
plt.show()
```

```python
fig_2, ax = plt.subplots(figsize=(15, 8))
im = ax.imshow(commit_hour, cmap='jet')
divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='5%', pad=0.05)
plt.colorbar(im, cax=cax)

ax.set_xticks(range(commit_hour.shape[1]))
ax.set_yticks(range(commit_hour.shape[0]))
ax.set_xticklabels(commit_hour.columns)
ax.set_yticklabels(commit_hour.index)

plt.setp(ax.get_xticklabels(), rotation=90)
plt.show()
```
